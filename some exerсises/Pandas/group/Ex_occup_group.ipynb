### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users=pd.read_csv(url, sep='|', index_col='user_id')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# first way
x=users.pivot_table(index='occupation',columns='gender', values='age', aggfunc='count').reset_index()
x=x.fillna(0)
x['total']=x['F']+x['M']
x['men_ratio']= round((x['M']/x['total']*100),2)
x.sort_values(['men_ratio'], ascending=False)

```

```python
# second way
def gender_to_numeric (x):
    if x=='M':
        return 1
    if x=='F':
        return 0
users['gender_n']=users['gender'].apply(gender_to_numeric)   
rat=(users.groupby('occupation').gender_n.sum()/users.groupby('occupation').gender_n.count()*100)
rat.sort_values( ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['max','min'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
round(users.pivot_table(index='occupation', columns='gender', values='age', aggfunc='mean'))
#users.groupby(['occupation', 'gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
per=users.pivot_table(index='occupation', columns='gender', values='age', aggfunc='count').fillna(0)
per['men_rat']=round((per['M']/(users.groupby('occupation').age.count())*100),2)
per['wom_rat']=round((per['F']/users.groupby('occupation').age.count()*100),2)
per
```
