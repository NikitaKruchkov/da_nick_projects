## Приоритизация гипотез.

```python
import pandas as pd
import scipy.stats as stats
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
import warnings
register_matplotlib_converters()
pd.set_option('display.max_colwidth', None)
```

```python
df=pd.read_csv('/datasets/hypothesis.csv')
```

```python
df
```

### Метод ICE

```python
# приоритезация по методу ICE
df['ICE']= round((df['Impact']*df['Confidence'])/df['Efforts'],2)

df_i=df.sort_values(by='ICE', ascending=False)
df_i.reset_index(drop=True, inplace=True)
df_i
```

### Метод RICE

```python
# приоритезация по методу ICE
df['RICE']=(df['Impact']*df['Confidence']*df['Reach'])/df['Efforts']

df_r=(df[['Hypothesis','Reach','Impact', 'Confidence','Efforts','ICE','RICE']]
      .sort_values(by=['RICE'], ascending=False))
df_r.reset_index(drop=True, inplace=True)
df_r
```

### Вывод

При использовании метода RICE в числитель к impact & confidence добавляется множитель охвата аудитории reach. Именно его присутствие в расчетах приводит к тому, что значение отличается от ICE. В данном кейсе я бы предпочёл использовать метод RICE, так как он предполагает наиболее полное использование параметров приоритизации гипотез, то есть при использовании ICE мы пренебрегаем параметром reach, что не соответствует принципу полноты информаци.
Просматривая таблицу с рейтингами гипотиз по RICE можно заметить, что есть гипотезы c одинаковой оценкой например гипотезы под индексом №2,№3. При ранжировании таких моментов я бы ориентировался на параметры confidence и efforts. В приоритет ставил бы более легковыполнимые гипотезы с большим уровнем увереноости в показателях. Если мой подход в ранжировании гипотез с одинаковой оценкой не верен, прошу дать комментарий, как поступать в таких ситуациях.

### Исследование кумулятивной выручки по группам.  График.

```python
#создаю табл. с уник. парами значений date/group.
dateGroups=orders[['date', 'group']].drop_duplicates()

#создаю таблицу ordersAgg содержит кол-во уник. платящих польз-ей, суммарную выручку по группам, 
#число уник.заказов по дату теста
ordersAgg=(dateGroups.apply(lambda x:orders[np.logical_and(orders['date']<=x['date'], orders['group']==x['group'])]
.agg({'date':'max', 'group':'max', 'visitorId':'nunique', 'transactionId':'nunique', 'revenue':'sum'}), axis=1)
.sort_values(by=['date', 'group']))
ordersAgg

```

```python
# создаю таблицу visitorsAgg с датой, группой и кол-ом уникальных пользователей по группам на опр дату.
visitorsAgg=(dateGroups.apply(lambda x:visitors[np.logical_and(visitors['date']<=x['date'], visitors['group']==x['group'])]
.agg({'date':'max', 'group':'max', 'visitors':'sum'}), axis=1)
.sort_values(by=['date', 'group']))
visitorsAgg
```

```python
# объединение таблиц и новые названия столбцов 
cumulData=ordersAgg.merge(visitorsAgg, left_on=['date','group'], right_on=['date','group'])
cumulData.columns=['date','group','buyers', 'orders','revenue', 'visitors']
cumulData.head(4)
```

```python
# создаю таблицы для разных групп с выручкой для постройки графиков
cumulRevA=cumulData[cumulData['group']=='A'][['date','revenue','orders']]
cumulRevB=cumulData[cumulData['group']=='B'][['date','revenue','orders']]
cumulRevB.head(5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(cumulRevA['date'], cumulRevA['revenue'], label='A')
plt.plot(cumulRevB['date'], cumulRevB['revenue'], label='B')
plt.xlabel('Дата')
plt.ylabel('Выручка, млн.руб.')
plt.xticks(rotation=90)
plt.legend()
plt.title('График куммулятивной выручки по группам')
```

# Вывод по графику кумулятивной выручки:
Исходя из графика видно, что за период наблюдения группа B принесла ощутимо большее кол-во денег. Но есть настораживающий момент, а  именно резкий пик в период с 18 по 19 августа. Он может быть вызван аномально большим заказом(-ми). Такой выброс искажает реальную картину.

### Исследование кумулятивного среднего чека по группам.  График.

```python
plt.figure(figsize=(10,6))
plt.xlabel('Дата')
plt.ylabel('Средний чек, р.')
plt.plot(cumulRevA['date'], cumulRevA['revenue']/cumulRevA['orders'], label="A")
plt.plot(cumulRevB['date'], cumulRevB['revenue']/cumulRevB['orders'], label="B")
plt.xticks(rotation=90)
plt.legend()
plt.title('Куммулятивный график среднего чека по группам')
```

#  Выводы по исследованию среднего чека:
Что и следовало ожидать. Снова скачок стоимости среднего чека в те же даты 18-20 августа. Невольно вспоминается анекдот про Билла Гейтса и бар. Действительно аномальные значения очень сильно влияют на среднюю и искажают картину, нужны дальнейшие исследования выбросов. По такому графику принять решение о выборе между A и B не будет верным решением.

### График относительного изменения кумулятивного среднего чека группы B к группе A. 

```python
#соединяю таблицы с кумул. выручкой для постройки графика относ. изменения ср.чека
mergCumulRev=cumulRevA.merge(cumulRevB, left_on=['date'], right_on=['date'], suffixes=['A','B'])
plt.figure(figsize=(10,6))
plt.xlabel('Дата')
plt.ylabel('Относительное колебание')
plt.plot(mergCumulRev['date'], 
(mergCumulRev['revenueB']/mergCumulRev['ordersB'])/(mergCumulRev['revenueA']/mergCumulRev['ordersA'])-1)
plt.axhline(y=0, color='black', linestyle='--')
plt.xticks(rotation=90)
plt.title('График колебаний среднего чека группы B относительно А')
```

# Вывод по графику относительного изменения кумулятивного среднего чека:
И снова граффик показывает, что средний чек группы В давал два пика относительно А- первый,  который я заметил ранее, в дату 18-19 авг, второй с 6-9 авг. В эти временные промежутки нужно искать аномально крупные продажи.

### График кумулятивного среднего количества заказов на посетителя по группам.

```python
# добавляю колонку конверсии в кумул. таблицу и разбиваю её по двум группам
cumulData['conversion']=cumulData['orders']/cumulData['visitors']
cumulData

cumulDataA=cumulData[cumulData['group']=="A"][['date','conversion']]
cumulDataB=cumulData[cumulData['group']=="B"][['date','conversion']]

```

```python
#нашел дату начала и конца наблюдений для изменения масштаба графика
cumulDataA['date'].min()
cumulDataA['date'].max()
plt.figure(figsize=(10,6))
plt.xlabel('Дата')
plt.ylabel('среднее кол-во заказов на посетителя')
plt.plot(cumulDataA['date'], cumulDataA['conversion'], label='A')
plt.plot(cumulDataB['date'], cumulDataB['conversion'], label='B')
plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), 0, 0.05])
plt.xticks(rotation=90)
plt.legend()
plt.title('Граффик среднего кол-ва заказов на посетителя')
```

# Вывод по графику среднего кол-ва закзаов на посетителя:
Из графика видно что в первые дни наблюдения группа А имела более высокое среднее количество заказов на посетителя, но с течением времени ситуация поменялась - В лидер. Повинуясь закону больших чисел показатели среднего устремились к своему среднему значению и к коцу наблюдения имели самое близкое к среднему значение, где среднее кол-во зак-ов В выше примерно на 10%.

### График относительного изменения кумулятивного среднего количества заказов на посетителя группы B к группе A. 

```python
mergcumulConv=cumulDataA.merge(cumulDataB, left_on=['date'], right_on='date', suffixes=['A','B'] )
plt.figure(figsize=(10,6))
plt.xlabel('Дата')
plt.ylabel('Относительное колебание')
plt.plot(mergcumulConv['date'], (mergcumulConv['conversionB']/mergcumulConv['conversionA'])-1)
plt.xticks(rotation=90)
plt.title('График относительного сравнения среднего кол-ва закзаов на посетителя группы В с А ')
plt.axhline(y=0, color='black', linestyle='--')
plt.axhline(y=0.13, color='black', linestyle='--')
plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), -0.5, 0.8])
```

# Выводы по графику:

### Считаю статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.

Уровень статистической значимости (alpha) определяю в 5%. Использую статистический критерий Манна-Уитни.

Нулевая гипотеза: Средние кол-ва заказов на посетителя в группе А и В равны.(сырые данные)

Альтернативная гипотеза: Средние кол-ва заказов на посетителя в группе А и В не равны.(сырые данные)

```python
# Изменю уже готовую аггрегац-ю таблицу по пользователям А, получу кумулят. сумму виситов на опр. дату visitorsADailyCml
# так же нужна таблица по кол-во посетителей А на дату visitorsADaily
visitorsADaily=visitors[visitors['group']=='A'][['date','visitors']]
visitorsADaily.columns=['date', 'visitorsPerDateA']

visitorsACml=visitorsAgg[visitorsAgg['group']=="A"][['date', 'visitors']].reset_index(drop=True)
visitorsACml.columns = ['date', 'visitorsCummulativeA']
visitorsACml.head(3)
```

```python
#такие же таблицы для группы В
visitorsBDaily=visitors[visitors['group']=='B'][['date','visitors']]
visitorsBDaily.columns=['date', 'visitorsPerDateB']

visitorsBCml=visitorsAgg[visitorsAgg['group']=="B"][['date', 'visitors']].reset_index(drop=True)
visitorsBCml.columns = ['date', 'visitorsCummulativeB']
visitorsBCml.head(3)
```

```python
# ordersACmlтаблица содержащая кумулятивные кол-во заказов и выручку на опр. дату
ordersACml=cumulData[cumulData['group']=="A"][['date', 'orders','revenue']]
ordersACml.columns = ['date','ordersCummulativeA','revenueCummulativeA']

ordersACml.head(5)

# ordersADaily таблица содерж-ая кол-во заказов и выручку на опр. датц
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'}))
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']
#ordersADaily.head(5)
#ordersACml.head(5)
```

```python
# такие же таблицы, но для В группы
ordersBCml=cumulData[cumulData['group']=="B"][['date', 'orders','revenue']]
ordersBCml.columns = ['date','ordersCummulativeB','revenueCummulativeB']

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'}))
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

```

```python
# слияние ранее собраных таблиц
data= (ordersADaily.merge(ordersBDaily, left_on='date', right_on='date', how='left')
       .merge(ordersACml, left_on='date', right_on='date', how='left')
       .merge(ordersBCml, left_on='date', right_on='date', how='left')
       .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
       .merge(visitorsADaily, left_on='date', right_on='date', how='left')
       .merge(visitorsACml, left_on='date', right_on='date', how='left')
       .merge(visitorsBCml, left_on='date', right_on='date', how='left')
      )
data.head(5)
```

```python
#создаю таблицы по группам, где будут указаны кол-во заказов по каждому платящему пользователю
ordersUsrsA=(orders[orders['group']=="A"]
             .groupby('visitorId', as_index=False)
             .agg({'transactionId':'nunique'}))
ordersUsrsA.columns=['visitorId', 'orders']


ordersUsrsB=(orders[orders['group']=="B"]
             .groupby('visitorId', as_index=False)
             .agg({'transactionId':'nunique'}))
ordersUsrsB.columns=['visitorId', 'orders']
ordersUsrsA.head(5)
```

```python
#создаю два датасета SampleA/B в которых каждому платящему клиенту будет соответствовать кол-во заказов, а не платящему ноль
sampleA=pd.concat(
    [ordersUsrsA['orders'],
     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum()-len(ordersUsrsA['orders'])), name='orders')], axis=0)

sampleB=pd.concat(
    [ordersUsrsB['orders'],
     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum()-len(ordersUsrsB['orders'])), name='orders')], axis=0)
sampleA
```

```python
# спомощью метода Манна-Уитни посчитаем p-value, для нулевой гипотезы о равенстве среднего кол-ва заказов по сырым данным
# уровень стат. значимости определил как 5%
alpha=0.05
p_value_o=stats.mannwhitneyu(sampleA, sampleB)[1]
if p_value_o<alpha:
    print('Не принимаю нулевую гипотезу о равенстве среднего кол-ва заказов')
else:
    print('Нулевая гипотеза равенства средних принимается')
    
# стоит проверить относительный прирост кол-ва закзаов группы В к группе А
print('{0:.3f}'.format(sampleB.mean()/sampleA.mean()-1))
```

по "сырым" данным нулевая гипотеза о равенстве среднего числа заказов на посетителя в группах А и Б не может быть принята. Этот вывод подтверждает так же прирост среднего по группе В в размере 13,8% относительно группы А. Надо сравнить этот вывод с тестом по очищенным данным.

Ген. совокупность нам неизвестна и мы оперируем только выборкой, в данной выборочной совокупности(группа В) выборочное среднее показало результат, гарантировано отличающийся от среднего ген совокупности, сохраняется вероятность, что при последующих наблюдениях сумма среднего чека будет колебаться. Плюс факт большего  на 15,1% кол-ва заказов на посетителя, даёт уверенность в большей доходности именно канала В.
