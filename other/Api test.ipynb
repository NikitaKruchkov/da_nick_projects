Параметры запроса:
1. `lat`- широта геогр-ой точки
2. `lon`- долгота геогр-ой точки
3. `lang`- язык ответа
4. `limit`- кол-во дней прогноза (max 7, по умолч.7)
5. `hours`- есть ли почасовой прогноз (по умолч. true)
6. `extra`- есть ли подробная инфа по осадкам (по умолч. true)
